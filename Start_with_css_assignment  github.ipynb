    </div>
</body>
</html>

2. Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using
class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and
explain how you've named the classes=

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>BEM Styled Paragraphs</title>
    <style>
        /*
        BEM Naming Convention:
        Block: A standalone entity (e.g., paragraph)
        Element: Part of a block (e.g., paragraph__title, paragraph__content)
        Modifier: Variation of a block or element (e.g., paragraph--primary)
        */

        /* Block: paragraph */
        .paragraph {
            margin-bottom: 15px; /* Space between paragraphs */
            padding: 10px;      /* Padding inside paragraph */
            border: 1px solid #ddd;  /* Light border */
            border-radius: 5px;
        }

        /* Element: paragraph__title */
        .paragraph__title {
            font-size: 1.2em;
            font-weight: bold;
            color: #333;
            margin-bottom: 5px;
        }

        /* Element: paragraph__content */
        .paragraph__content {
            line-height: 1.5;    /* Increased line height for readability */
            color: #555;
        }

        /* Modifier: paragraph--primary */
        .paragraph--primary {
            background-color: #f0f4c3; /* Light yellow background */
            border-color: #d4e157;     /* Darker yellow border */
            color: #222;
        }

        /* Modifier: paragraph--secondary */
        .paragraph--secondary {
            background-color: #e1f5fe; /* Light cyan */
            border-color: #81d4fa;
            color: #222;
        }

        /* Modifier: paragraph--highlighted */
        .paragraph--highlighted {
            background-color: #fff3e0; /* Light orange */
            border-color: #ffb74d;
            color: #222;
        }

        .paragraph--highlighted .paragraph__title{
             color: #e65100; /* Darker orange for title in highlighted para */
        }
    </style>
</head>
<body>
    <div class="paragraph paragraph--primary">
        <h2 class="paragraph__title">Primary Paragraph</h2>
        <p class="paragraph__content">This is the primary paragraph.  It has a light yellow background.</p>
    </div>

    <div class="paragraph paragraph--secondary">
        <h2 class="paragraph__title">Secondary Paragraph</h2>
        <p class="paragraph__content">This is the secondary paragraph. It has a light cyan background.</p>
    </div>

    <div class="paragraph paragraph--highlighted">
        <h2 class="paragraph__title">Highlighted Paragraph</h2>
        <p class="paragraph__content">This paragraph is highlighted. It has a light orange background.</p>
    </div>

    <div class="paragraph">
        <h2 class="paragraph__title">Default Paragraph</h2>
        <p class="paragraph__content">This is a default paragraph.  It has a light grey border.</p>
    </div>
</body>
</html>

3. Develop an HTML form with various input elements. Use CSS to style the form, including setting
background colors for input fields. Create a custom color palette for the form elements, and demonstrate
how to apply opacity to one of the form sections.

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Styled Form</title>
    <style>
        /* Custom Color Palette */
        :root {
            --primary-color: #a7c957; /* Light Green */
            --secondary-color: #f2e9e4; /* Off-White */
            --accent-color: #6a4c93; /* Purple */
            --text-color: #3d405b; /* Dark Gray */
        }

        body {
            font-family: sans-serif;
            background-color: var(--secondary-color);
            display: flex;
            justify-content: center;
            align-items: center;
            min-height: 100vh;
            margin: 0;
        }

        .form-container {
            background-color: white;
            padding: 30px;
            border-radius: 8px;
            box-shadow: 0 0 20px rgba(0, 0, 0, 0.1);
            width: 400px;
        }

        .form-group {
            margin-bottom: 20px;
        }

        label {
            display: block;
            margin-bottom: 5px;
            color: var(--text-color);
            font-weight: bold;
        }

        input[type="text"],
        input[type="email"],
        input[type="password"],
        textarea,
        select {
            width: calc(100% - 12px);
            padding: 10px;
            border: 1px solid #ccc;
            border-radius: 4px;
            font-size: 16px;
            background-color: var(--secondary-color); /* Applying a color from the palette */
            color: var(--text-color);
            box-sizing: border-box; /* Ensure padding doesn't affect width */
        }

        textarea {
            resize: vertical;
        }

        button {
            background-color: var(--primary-color);
            color: white;
            padding: 12px 20px;
            border: none;
            border-radius: 4px;
            cursor: pointer;
            font-size: 16px;
            transition: background-color 0.3s ease;
        }

        button:hover {
            background-color: #81a146;
        }

            background-color: #f9f9f9;

            border-radius: 4px;

            opacity: 0.8; /* Applying opacity */
        }

        }

ey features and explanations:

Custom Color Palette:
CSS variables (defined within the :root pseudo-class) are used to store a set of reusable colors:
--primary-color (light green)
--secondary-color (off-white)
--accent-color (purple)
--text-color (dark gray)
This approach promotes consistency and simplifies color scheme management. To use a color, we use var(--variable-name).
Form Styling:
The .form-container provides a styled container for the form, with a white background, padding, rounded corners, and a subtle shadow.
.form-group adds vertical spacing between form element groups (label and input).
Labels are styled with display: block to ensure they appear above their corresponding input fields, and are given a dark gray color and bold font weight.
Input fields (input[type="text"], input[type="email"], etc.), textarea, and select elements are styled with:
Full width (minus padding)
Padding for better visual appearance
A light gray border with rounded corners
The off-white background color from the custom palette
Dark gray text color
box-sizing: border-box to include padding and border in the element's total width
textarea is made resizable only in the vertical direction using resize: vertical;.
The submit button is styled with:
The primary color from the palette (light green)
White text
Padding
No border, and rounded corners
A pointer cursor for better user interaction
A subtle transition for a hover effect
The button:hover pseudo-class changes the button's background color when the user hovers over it.
Opacity:
The .contact-info section has opacity: 0.8; applied, making it slightly transparent. This can be used to de-emphasize less critical information or create a layered effect.
